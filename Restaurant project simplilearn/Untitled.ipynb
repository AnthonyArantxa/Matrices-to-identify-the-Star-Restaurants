```python
split_list=[]
for i in range(9541):
    cuisine=cuisine_list[i].split(",")
    split_list.append(cuisine)
```

```python
final_list=[]
for i in range(9541):
    l=len(split_list[i])
    element=split_list[i]
    for j in range(l):
        final_list.append(element[j].strip(" "))
```

```python
set_list=set(final_list)
Cuisines=list(set_list)
```

```python
Frequency=[]
for i in Cuisines:
    count=final_list.count(i)
    Frequency.append(count)
    print(i, count)
```

```python
Top_cuisines= pd.DataFrame(list(zip(Cuisines, Frequency)),
               columns =['Cuisines', 'Frequency'])
Top_cuisines
```

```python
data["Exchange Rate"]=ex_rate
data["Country Name"]=Country_name
```
